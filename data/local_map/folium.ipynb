```python
import pandas as pd

file_path = '../meteorological/location.csv'
location_data = pd.read_csv(file_path)

import folium

center_lat = location_data['위도'].mean()
center_lon = location_data['경도'].mean()
m = folium.Map(location=[center_lat, center_lon], zoom_start=8)

for _, row in location_data.iterrows():
    folium.Marker(
        location=[row['위도'], row['경도']],
        popup=f"{row['(현)지점명']}<br>{row['(현)주소']}",
        tooltip=row['(현)지점명']
    ).add_to(m)

m
```
