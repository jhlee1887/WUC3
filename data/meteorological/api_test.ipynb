```python
import requests
import json
import os
from datetime import datetime

# 설정
api_url = "	http://apis.data.go.kr/1360000/AsosHourlyInfoService/getWthrDataList"
api_key = "sj40PraKG7dGcBNarkyuQo1EcwXDwDORJi1Hz103NVmLD4NBFkyIHBiEnGw/XnL8pKczzFOJh5+qlCtxwRCBDw=="  # 적절한 API 키로 대체하세요.
save_dir = "./retry_results"
os.makedirs(save_dir, exist_ok=True)

# 호출하려는 오류 데이터
station_id = 123  # 스테이션 ID를 수정하세요.
start_dt = "19720101"  # 시작 날짜를 수정하세요.
end_dt = "19720210"  # 종료 날짜를 수정하세요.

# 요청 파라미터
params = {
    "authKey": api_key,
    "pageNo": "1",
    "numOfRows": "999",
    "dataType": "JSON",
    "dataCd": "ASOS",
    "dateCd": "HR",
    "startDt": start_dt,
    "startHh": "00",
    "endDt": end_dt,
    "endHh": "23",
    "stnIds": station_id
}

try:
    response = requests.get(api_url, params=params, timeout=10)
    if response.status_code == 200:
        try:
            data = response.json()
            file_path = os.path.join(save_dir, f"station_{station_id}_{start_dt}_{end_dt}.json")
            with open(file_path, "w", encoding="utf-8") as f:
                json.dump(data, f, ensure_ascii=False, indent=4)
            print(f"Data saved to {file_path}")
        except json.JSONDecodeError:
            file_path = os.path.join(save_dir, f"station_{station_id}_{start_dt}_{end_dt}_raw.txt")
            with open(file_path, "w", encoding="utf-8") as f:
                f.write(response.text)
            print(f"Raw response saved to {file_path}")
    else:
        print(f"HTTP Error {response.status_code}: {response.text}")
except requests.exceptions.RequestException as e:
    print(f"Request failed: {e}")
```
