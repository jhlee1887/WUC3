```python
import pandas as pd

file_path = './kosis_data.csv'
data = pd.read_csv(file_path)

area_data = data[data['ITM_NM'] == '면적']
area_pivot = area_data.pivot_table(
    index=['PRD_DE', 'C1_NM'],
    columns='C2_NM',
    values='DT',
    aggfunc='first'
).reset_index()

area_pivot.rename(columns={'도시지역': '도시지역 면적 (㎡)', '녹지지역': '녹지지역 면적 (㎡)'}, inplace=True)

ratio_data = data[data['ITM_NM'] == '비율']
merged_data = pd.merge(
    area_pivot,
    ratio_data[['PRD_DE', 'C1_NM', 'DT']].rename(columns={'DT': '녹지율 (%)'}),
    on=['PRD_DE', 'C1_NM'],
    how='left'
)

output_file_path = './kosis_dataframe.csv'
merged_data.to_csv(output_file_path, index=False, encoding='utf-8-sig')
```
